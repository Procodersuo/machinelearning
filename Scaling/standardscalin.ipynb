```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("E:\ml\Scaling\Social_Network_Ads.csv")
df=df.iloc[:,2:]
df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),df['Purchased'], random_state=0, test_size=0.3)
X_train
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
temp=pd.DataFrame(X_train_scaled)

```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr_scaled=LogisticRegression()
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
prediction=lr.predict(X_test)
predictionscaled=lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction))
print(accuracy_score(y_test,predictionscaled))
```
