```python
import math
myData=[27,28,31,24.89,55,53,53,242,52,23,24,242,52,22,252,52]
sumofdata=0
mean=0
for i in myData:
    sumofdata=sumofdata+i
mean=sumofdata/(len(myData))
squareDiff=[]
for i in myData:
    squareDiff1=i-mean
    squareDiff.append(squareDiff1**2)
squareDiffSum=0
for i in squareDiff:
    squareDiffSum=squareDiffSum+i
squareDiffSum=squareDiffSum/(len(myData)-1)
math.sqrt(squareDiffSum)

```
